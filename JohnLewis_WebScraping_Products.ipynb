```python
previous_height = 0
chunk_increment = 1
print(chunk_increment)

while True:
    url_test = 'https://www.johnlewis.com/browse/electricals/coffee-machines/view-all-coffee-machines/_/N-afu?chunk='+str(chunk_increment)

    driver = webdriver.Chrome('/tmp/chromedriver')
    driver.get(url_test)
    time.sleep(2)
    driver.find_element(By.XPATH,'//*[@data-test="allow-all"]').click()
    new_height = driver.execute_script("return document.documentElement.scrollHeight")
    print(chunk_increment,new_height,previous_height, driver.execute_script("return document.documentElement.scrollHeight"))
    if new_height > previous_height:
        previous_height = new_height
        chunk_increment += 1
        time.sleep(2)
    else:
        break
```

```python
html = BeautifulSoup(driver.page_source)
html
```

```python
session = HTMLSession()
def getdata(url):
    get_url = session.get(url)
    page_url = BeautifulSoup(get_url.text, 'html.parser')
    return page_url
```

```python
allproducts = html.find_all(attrs={'class':'product-card_c-product-card__UAdsG ProductGrid_product__ER835'})
```

```python
gc.collect()
product_details = []
for product in range(0, len(allproducts)):
    product_id = ['p'+[allproducts[product].find(attrs={'class':'image_imageLink__1Znsz product-card_c-product-card__image__bO3kW product__image'}).get('href')][0].split('/p')[-1] if allproducts[product].find(attrs={'class':'image_imageLink__1Znsz product-card_c-product-card__image__bO3kW product__image'}).get('href') else 'N/A'][0]
    brand = [allproducts[product].find(attrs = {'class':'title_title__brand__UX8j9 undefined'}).text if allproducts[product].find(attrs = {'class':'title_title__brand__UX8j9 undefined'}) else 'N/A'][0]
    product_name = [allproducts[product].find(attrs = {'title_title__desc__ZCdyp title_title__desc--three-lines__VHz1t title_title__desc--branded__8SluU'}).text if allproducts[product].find(attrs = {'title_title__desc__ZCdyp title_title__desc--three-lines__VHz1t title_title__desc--branded__8SluU'}) else 'N/A'][0]
    price = [allproducts[product].find(attrs={'class':'price_price__now__3B4yM'}).text if allproducts[product].find(attrs={'class':'price_price__now__3B4yM'}) else 'N/A'][0]
    ratings = [allproducts[product].find(attrs={'class':'rating_visuallyHidden__zgQQN'}).text.split(',')[3].strip().split(' ')[0] if allproducts[product].find(attrs={'class':'rating_visuallyHidden__zgQQN'}) else 'N/A'][0]
    reviews = [allproducts[product].find(attrs={'class':'rating_suffix__PdYl6 rating_suffix-joyfully-bold__XnIse'}).text.split( )[2] if allproducts[product].find(attrs={'class':'rating_suffix__PdYl6 rating_suffix-joyfully-bold__XnIse'}) else 'N/A'][0]
    promotions = [allproducts[product].find(attrs={'data-test':'product-card-promo-messages'}).text if allproducts[product].find(attrs={'data-test':'product-card-promo-messages'}) else 'N/A'][0] #promotions
    href = ['https://www.johnlewis.com'+allproducts[product].find(attrs={'class':'image_imageLink__1Znsz product-card_c-product-card__image__bO3kW product__image'}).get('href') if allproducts[product].find(attrs={'class':'image_imageLink__1Znsz product-card_c-product-card__image__bO3kW product__image'}).get('href') else 'N/A'][0]
    print(product_id, product, brand, product_name, price, ratings, reviews, promotions,href)
    product_details.append({'Product_ID':product_id,'Brand':brand,'Product_name':product_name, 'Price':price,'Ratings':ratings,'Reviews':reviews, 'Promotions':promotions, 'Product_link': href})
```

```python
df = pd.DataFrame.from_dict(product_details)
df
df.to_csv('product_details.csv', encoding='utf-8-sig', index=False)
gc.collect()
```

```python
product_specifications = []
for each_prod in range(0,len(product_details)):
    time.sleep(2)
    each_product_info = getdata(product_details[each_prod]['Product_link'])
    product_specif = each_product_info.find_all(attrs={'class':'ProductSpecification_productSpecificationListItem__2Za7A'})
    temp_val = {'Product_ID':product_details[each_prod]['Product_ID']}
    for each_prod_spec in range(0,len(product_specif)):
        temp = [product_specif[each_prod_spec].find(attrs={'data-testid':'product:specification:list:value'}).text if product_specif[each_prod_spec].find(attrs={'data-testid':'product:specification:list:label'}).text else 'N/A'][0]
        if temp != 'N/A':
            temp_val[product_specif[each_prod_spec].find(attrs={'data-testid':'product:specification:list:label'}).text] = temp
            print(temp_val)
    product_specifications.append(temp_val)
        
```
